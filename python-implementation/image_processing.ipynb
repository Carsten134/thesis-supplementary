```python
import matplotlib.pyplot as plt
from PIL import Image
from torchvision.transforms.functional import pil_to_tensor
from source import phi_n_star
```

```python
def do_test(image_1, image_2):
  # fetch images
  x = pil_to_tensor(Image.open(f"./data/{image_1}"))
  y = pil_to_tensor(Image.open(f"./data/{image_2}"))
  
  # normalizing the image
  x = x.sum(0)/ 450*600
  y = y.sum(0)/ 450*600
  
  # do test and print result
  decision, Tn, Tn_star = phi_n_star(x, y, 500, .06, stride = 10, hr = .005, hc=.005)
  plt.hist(Tn_star.reshape(1,500)[0], bins = 20)
  plt.title("Distribution of Tn_star")
  plt.vlines(Tn, 0, 100, colors = ["red"])
  plt.show()
  print(decision)
```

```python
do_test("car.png", "ISIC_0029321.jpg")
```
