## Plotting H0 Behavior For Equality Teststat
1. Concatinate all simulation results
2. Aggregate with the mean
3. Visualise with python utils

```python
import pandas as pd

df_ma = pd.read_csv("./sim_results/size/with_symmetry/mc_band_ma.csv")
df_ma_row = pd.read_csv("./sim_results/size/with_symmetry/mc_band_ma_row.csv")
df_ar = pd.read_csv("./sim_results/size/with_symmetry/mc_band_ar.csv")
df_ma_cor = pd.read_csv("./sim_results/size/with_symmetry/mc_band_ma_corr.csv")
```

```python
# aggregration
rej_quarter = df_ar.loc[:, ["h", "decision"]].groupby("h").mean().decision
rej_row = df_ma_row.loc[:, ["h", "decision"]].groupby("h").mean().decision
rej_full = df_ma.loc[:, ["h", "decision"]].groupby("h").mean().decision
rej_cor = df_ma_cor.loc[:, ["h", "decision"]].groupby("h").mean().decision

hs = df_ma.h.unique()
```

```python
import matplotlib.pyplot as plt
import numpy as np # Imported just to generate dummy data for the example

# --- 1. Define the Professional Style ---
# Use a cleaner base style
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
    'axes.labelsize': 14,               # larger axis labels
    'axes.titlesize': 16,               # larger title
    'xtick.labelsize': 12,              # larger tick labels
    'ytick.labelsize': 12,
    'legend.fontsize': 12,
    'figure.figsize': (8, 5),           # Standard ratio for papers
    'lines.linewidth': 2.0   
})

# --- 2. Plotting ---
# Use 300 DPI for print quality
fig, ax = plt.subplots(dpi=300)

# Use markers and distinct linestyles for B&W compatibility
ax.plot(hs, rej_quarter, label="$QP$", color='#333333', linestyle='-', marker='o', markersize=5)
ax.plot(hs, rej_row, label="$MA_{row}$", color='#666666', linestyle='--', marker='s', markersize=5)
ax.plot(hs, rej_full, label="$MA_{full}$", color='#999999', linestyle=':', marker='^', markersize=6)
ax.plot(hs, rej_cor, label="$MA_{corr}$", color="#8B8B8B", linestyle='--', marker='o', markersize=6)

# --- 3. Refinements ---
# Use axhline for the threshold line (it automatically spans the whole plot)
ax.axhline(y=0.05, color='firebrick', linestyle='--', linewidth=1.5, alpha=0.8, label=r'$\alpha = 0.05$')

# Labels
# ax.set_title("R $H_0$ (N=1000, 20x20 Gitter)", pad=15) # Using LaTeX for H0
ax.set_ylabel("Nominal Size")
ax.set_xlabel("Kernel Bandwidth")

# Grid and Legend
ax.grid(True, which='major', linestyle='-', alpha=0.6)
ax.legend(title="Process", frameon=True, framealpha=1, facecolor='white', edgecolor='black')

# Tight layout handles padding automatically
plt.tight_layout()

# Save as PDF for LaTeX/Word (Best practice)
# plt.savefig("academic_plot.pdf", format='pdf', bbox_inches='tight')

plt.show()
```
